Example Output

Recursion means a function calling itself to solve a smaller version of the same problem. Think of it like opening a box and finding another smaller box inside until you finally reach the smallest one. That stopping point is called the base case.

Example: factorial(5) = 5 × factorial(4). The function keeps calling itself until factorial(1).
