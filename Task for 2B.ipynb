Task for 2B

```python
import json
import re

txt_file = "10_mcq_questions_with_answers.txt"

with open(txt_file, "r", encoding="utf-8") as f:
    content = f.read()

blocks = re.split(r'\n(?=\d+\.\s)', content.strip())

data = []

for block in blocks:

    lines = [line.strip() for line in block.splitlines() if line.strip()]

    first = re.match(r'(\d+)\.\s+(.*)', lines[0])

    qid = int(first.group(1))
    question = first.group(2)

    options = {}
    answer = ""

    for line in lines[1:]:

        option_match = re.match(r'([A-D])\.\s+(.*)', line)
        answer_match = re.match(r'Answer:\s*([A-D])', line)

        if option_match:
            options[option_match.group(1)] = option_match.group(2)

        if answer_match:
            answer = answer_match.group(1)

    data.append({
        "id": qid,
        "question": question,
        "options": options,
        "answer": answer
    })

json_file = "mcq_questions.json"

with open(json_file, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2)

print("JSON file created successfully.")
print("Saved as:", json_file)
```
